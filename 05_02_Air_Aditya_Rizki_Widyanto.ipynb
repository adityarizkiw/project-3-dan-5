### User-defined Functions (UDFs)

- To print total model parameters

```python
def check_params(model, *args, **kwargs) -> dict:
    return {
        "total_trainable_params" : sum(p.numel() for p in model.parameters() if p.requires_grad),
        "total_nontrainable_params" : sum(p.numel() for p in model.parameters() if not p.requires_grad)
    }
```

- To get the pair of train and validation dataloaders

```python
data_transform = torchvision.transforms.Compose([
    torchvision.transforms.Resize((224, 224)),
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize((torch.tensor(33.3184)/255,), (torch.tensor(78.5675)/255,))
])

train_dataset = torchvision.datasets.MNIST(root=".", train=True, transform=data_transform, download=True).train_data.float()

def get_dataloaders(train_batch_size : int, val_batch_size : int, max_rows : int = 1000, *args, **kwargs) -> tuple:
    data_transform = torchvision.transforms.Compose([
        torchvision.transforms.Resize((224, 224)),
        torchvision.transforms.ToTensor(),
        torchvision.transforms.Normalize((torch.tensor(33.3184)/255,), (torch.tensor(78.5675)/255,))
    ])

    train_dataset = torchvision.datasets.MNIST(root=".", train=True, transform=data_transform)
    train_idx = torch.randperm(len(train_dataset))[:int(max_rows*.75)]
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=train_batch_size, \
                                               sampler=torch.utils.data.SubsetRandomSampler(train_idx))

    val_dataset = torchvision.datasets.MNIST(root=".", train=False, transform=data_transform)
    val_idx = torch.randperm(len(val_dataset))[:int(max_rows*.25)]
    val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=val_batch_size, \
                                             sampler=torch.utils.data.SubsetRandomSampler(val_idx))

    return train_loader, val_loader
```

* To fit (training) the model

```python
def fit(
    model : torchvision.models,
    epoch : int,
    train_loader : torch.utils.data.DataLoader,
    val_loader : torch.utils.data.DataLoader,
    *args, **kwargs
) -> dict:

    TRAIN_LOSS, TRAIN_ACC = [], []
    train_batches = len(train_loader)

    VAL_LOSS, VAL_ACC = [], []
    val_batches = len(val_loader)

    # loop for every epoch (training + evaluation)
    start_ts = time.time()
    for e in range(epoch):
        train_losses = 0
        train_accuracies = 0

        # progress bar
        progress = tqdm(enumerate(train_loader), desc="Loss: ", total=train_batches)

        # ----------------- TRAINING  --------------------
        # set model to training
        model.train()

        for i, data in progress:
            X, y = data[0].to(device), data[1].to(device)

            # training step for single batch
            model.zero_grad()

            # forward pass
            outputs = model(X)
            loss = loss_function(outputs, y)

            # backward pass
            loss.backward()
            optimizer.step()

            train_losses += loss.item()

            ps = torch.exp(outputs)
            top_p, top_class = ps.topk(1, dim=1)
            equals = top_class == y.view(*top_class.shape)
            train_accuracies += torch.mean(equals.type(torch.FloatTensor)).item()

            # updating progress bar
            progress.set_description("Loss: {:.4f}".format(train_losses/(i+1)))

        TRAIN_ACC.append(train_accuracies/train_batches)
        TRAIN_LOSS.append(train_losses/train_batches)

        # releasing unceseccary memory in GPU
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

        # ----------------- VALIDATION  -----------------
        val_losses = 0
        val_accuracies = 0

        # set model to evaluating (testing)
        model.eval()
        with torch.no_grad():
            for i, data in enumerate(val_loader):
                X, y = data[0].to(device), data[1].to(device)
                outputs = model(X) # this gives the prediction from the network
                val_losses += loss_function(outputs, y).item()

                ps = torch.exp(outputs)
                top_p, top_class = ps.topk(1, dim=1)
                equals = top_class == y.view(*top_class.shape)
                val_accuracies += torch.mean(equals.type(torch.FloatTensor)).item()

        print("Epoch {}/{} >> Training loss: {:.3f}, Validation loss: {:.3f}, Validation accuracy: {:.3f}".format(
            e+1, epoch, train_losses/train_batches, val_losses/val_batches, val_accuracies/val_batches*100)
        )

        VAL_ACC.append(val_accuracies/val_batches)
        VAL_LOSS.append(val_losses/val_batches)

    tr_time = time.time()-start_ts
    print("Training time: {:.3f}s".format(tr_time))

    return {
        "model" : model.name,
        "train_acc" : TRAIN_ACC,
        "train_loss" : TRAIN_LOSS,
        "val_acc" : VAL_ACC,
        "val_loss" : VAL_LOSS,
        "exc_time" : tr_time
    }
```

* To visualize the model performance

```python
def plot_performance(dict_ : dict, *args, **kwargs) -> None:
    my_figure = plt.figure(figsize=(12, 4))
    # NOTE: figsize=(width/horizontally, height/vertically)

    m = my_figure.add_subplot(121)
    plt.plot(dict_["train_loss"], label="Train Loss")
    plt.plot(dict_["val_loss"], label="Valid. Loss")
    plt.title("LOSS")
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.legend(loc="best")

    n = my_figure.add_subplot(122)
    plt.plot(dict_["train_acc"], label="Train Accuracy")
    plt.plot(dict_["val_acc"], label="Valid. Accuracy")
    plt.title("ACCURACY")
    plt.xlabel("Epoch")
    plt.ylabel("Score")
    plt.legend(loc="best")

    plt.tight_layout()
    plt.show()
```

### Define the model class

```python
class VisionModel(torch.nn.Module):
    def __init__(self, model_selection : str, *args, **kwargs) -> None:
        super(VisionModel, self).__init__()
        self.model_selection = self.name = model_selection
        self.in_channels = 1

        def create_conv2d(this_layer, *args, **kwargs) -> torch.nn.modules.conv.Conv2d:
            return torch.nn.Conv2d(
                in_channels=self.in_channels, out_channels=this_layer.out_channels,
                kernel_size=this_layer.kernel_size, stride=this_layer.stride,
                padding=this_layer.padding, bias=this_layer.bias
            )

        if not self.model_selection.lower() in ["resnet", "densenet", "vit"]:
            raise ValueError("Please select the model: 'resnet', 'densenet', or 'vit'.")

        if self.model_selection == "resnet":
            self.model = torchvision.models.resnet18(pretrained=True)
            self.model.conv1 = create_conv2d(self.model.conv1) # change the input layer to take Grayscale image, instead of RGB
            self.model.fc = torch.nn.Linear(self.model.fc.in_features, 10) # change the output layer to output 10 classes

        elif self.model_selection == "densenet":
            self.model = torchvision.models.densenet121(pretrained=True)
            self.model.features.conv0 = create_conv2d(self.model.features.conv0) # TODO: Change the DenseNet input layer stack by calling create_conv2d()
            self.model.classifier = torch.nn.Linear(self.model.classifier.in_features, 10) # TODO: Change the DenseNet output layer with 10 classes

        elif self.model_selection == "vit":
            self.model = torchvision.models.vit_b_16(pretrained=True)
            self.model.conv_proj = create_conv2d(self.model.conv_proj) # TODO: Change the ViT input layer stack by calling create_conv2d()
            self.model.classifier = torch.nn.Linear(self.model.classifier.in_features, 10) # TODO: Change the ViT output layer with 10 classes

        self.softmax = torch.nn.Softmax(dim=1)

    def forward(self, data, *args, **kwargs) -> torchvision.models:
        x = self.model(data)
        return self.softmax(x)
```

[ ANSWER HERE ] <br>
- Lack of Adaptation: When you freeze the layers of a pre-trained model, you're essentially using those layers as feature extractors. While this can be very effective when the features learned by the pre-trained model are relevant to your specific task, it can be detrimental if the features are not well-suited to the new task. The frozen layers may not adapt to the specific nuances and patterns in the new dataset, resulting in suboptimal performance.

- Task Dissimilarity: If the task you are transferring the model to is significantly different from the task the pre-trained model was originally trained on, then frozen layers may not be able to capture the necessary patterns and features for the new task. In such cases, fine-tuning some or all of the layers may be necessary to allow the model to adapt to the new data and task.

- Over-regularization: Freezing layers effectively prevents any updates to the weights in those layers during training. This can lead to over-regularization, where the model becomes too rigid and is unable to adapt to the specifics of the new task. Fine-tuning allows for some degree of adaptation and adjustment, which can be crucial for achieving higher accuracy.

- Underfitting: With a limited number of training epochs, frozen layers may not have enough opportunities to adapt to the new data and task. If the frozen layers are not a good match for the new task, they may underperform, and fine-tuning could help address this issue.

- Dataset Size: If your dataset is small, freezing layers might lead to worse accuracy because the model is more likely to overfit to the limited amount of data, especially if the features learned by the pre-trained model are not well-suited to the new task. Fine-tuning can help mitigate overfitting and adapt the model to the new task.

[ ANSWER HERE ] <br>
- Loss of Task Relevance: The layers in a pre-trained model have learned representations specific to the original task they were trained on. If you freeze a substantial number of these layers, they may not be well-aligned with the new task you're training the model for. As a result, the model's early predictions might be less relevant or informative for the new task, causing a drop in accuracy.

- Feature Extraction: The layers that are frozen act as feature extractors. If these frozen features don't capture relevant information for your specific task, the model might struggle to make meaningful predictions early on. Fine-tuning these layers allows the model to adapt to the new task by adjusting its feature representations.

- Gradient Flow: During training, gradients are backpropagated from the loss function through the network to update the model's weights. When many layers are frozen, the gradients may not flow effectively through the frozen layers to reach the trainable layers. This can result in slower convergence or difficulties in training the model.

- Overfitting Concerns: With fewer layers to update, the model has a reduced capacity to adapt to the new task. In early epochs, this can lead to overfitting because the model may quickly memorize the training data rather than learning meaningful patterns. Fine-tuning more layers can help the model generalize better.

[ ANSWER HERE ] <br>
- Fewer Parameters to Update: The frozen layers have a fixed set of parameters that don't change during training. Since updating parameters is a computationally expensive process, having fewer parameters to update means fewer computations per training step, resulting in faster training times.

- Less Backpropagation: During backpropagation, gradients are computed and used to adjust the model's weights. When more layers are frozen, there are fewer layers that need to be considered during the backpropagation process, which reduces the computational cost and speeds up the training process.

- Memory Efficiency: Fewer parameters and layers to update also reduce the memory requirements during training. This can be beneficial, especially when working with limited GPU or CPU memory, as you can fit larger batch sizes into memory, which further accelerates training.

- Stable Representations: Frozen layers are typically responsible for extracting high-level features that are less likely to change significantly for a related task. By keeping these layers fixed, you maintain stable and useful representations, which can help in the early stages of training and reduce the need for extensive updates.
