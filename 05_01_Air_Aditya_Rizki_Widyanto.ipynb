### Import libraries

``````python
import numpy as np, cv2
from skimage.io import imsave
from skimage.measure import block_reduce
from torch.nn import MaxPool2d
from torch import from_numpy
from matplotlib import pyplot as plt
from matplotlib.image import imread
``````

### User-defined Functions (UDFs)

``````python
def plot(
    imgs : list,
    titles : list = ["ORIGINAL", "EDITED"],
    figsize : tuple = (8, 5),
    cmaps : list = [None, None],
    *args, **kwargs
) -> None:

    if len(imgs) > 3:
        raise ValueError("Max. total image = 3.")

    if len(imgs) != len(titles):
        raise ValueError("'imgs' should be in the same length with 'titles'.")

    my_figure = plt.figure(figsize=figsize)
    # NOTE: figsize=(width/horizontally, height/vertically)

    fig_ids = [121, 122] if len(imgs) == 2 else [131, 132, 133]

    for idx, img, title, c in zip(fig_ids, imgs, titles, cmaps):
        m = my_figure.add_subplot(idx)
        plt.imshow(img, cmap=c)
        plt.axis("off")
        plt.title(title)

    plt.tight_layout()
``````

``````python
def apply_clahe_rgb(img, clipLimit=2, tileGridSize=(10, 10)):
    # LAB color model where L = Lightness, a = green-red, b = blue-yellow
    light, a, b = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2LAB)) # convert to LAB color model
    light_clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize).apply(light) # apply CLAHE for lightness (L) component only
    return cv2.cvtColor(cv2.merge((light_clahe, a, b)), cv2.COLOR_LAB2BGR) # return with reconvert to BGR color model
``````

### Max Pooling (using Scikit-image)

``````python
BLOCK_SIZE = (4, 4, 1)
``````

``````python
img_edited = block_reduce(img, block_size=BLOCK_SIZE, func=np.max)
plot(imgs=[img, img_edited])
``````

### Max Pooling (using PyTorch)

``````python
# convert numpy-based image to torch.Tensor type
img_torch = from_numpy(img).permute(2, 0, 1).float()

# prepare Max Pooling operation from PyTorch
max_pooling = MaxPool2d(kernel_size=9, stride=2)

# perform Max Pooling operation
img_torch_edited = max_pooling(img_torch).permute(1, 2, 0).int().numpy()
plot(imgs=[img, img_torch_edited])
``````

[ANSWER HERE] <br>
PyTorch's max pooling operation is built into the deep learning framework and is best suited for deep learning tasks where GPU acceleration and automatic differentiation are important. Scikit image max pooling, on the other hand,  is designed for general image processing and offers more flexibility regarding kernel shape, making it a valuable tool for many segmentation tasks. image analysis without deep learning. Choosing between the two depends on the specific requirements and context of your image processing project

[ ANSWER HERE ] <br>
The main difference between min pooling and average pooling is what is done on the pixel values within the pooling window. The minimum group selects the minimum value, emphasizing dark features, while the average pooling calculates the average, providing a smoother, less detailed rendering of the image.
If you don't see a clear difference between min pooling and average pooling for a particular image or block size, it may be due to image characteristics  and  pooling window size. You can experiment with different images or adjust the block size to see more pronounced differences in clustering results, especially in images with different contrast and intensity.

### CLAHE
Now imagine you have a better approach to compete Apple's Photonic Engine and Samsung's Adaptive Tetra-squared Pixel technologies using **Contrast Limited Adaptive Histogram Equalization (CLAHE)**. How's the result?

``````python
img_clahe = apply_clahe_rgb(img) # TODO: Call the function apply_clahe_rgb() and pass the variable image of "photo1.jpeg"
plot(imgs=[img, img_edited, img_clahe], titles=["ORIGINAL", "MAX-POOLING", "CLAHE"], figsize=(12, 5), cmaps=[None]*3)
``````

``````python
# QUESTIONS
# TODO: What are the advantages of using CLAHE (instead of Max Pooling) to brighten the dark-toned image?
``````

[ ANSWER HERE ] <br>

- Local Contrast Enhancement: CLAHE operates at the pixel level and can enhance local contrast in an image, especially in areas that are too dark or too bright. This means that CLAHE can help clarify important details in the image that might be lost if using Max Pooling alone.

- intensity histograms for small blocks within the image and then applies histogram equalization techniques to these blocks. This allows CLAHE to automatically adjust the brightness levels in areas with different intensities.

- Contrast Limit Control: CLAHE also has a "contrast limit" parameter that allows you to control the extent to which contrast is enhanced in the image. This can help prevent excessive contrast enhancement, which may occur when using Max Pooling alone.

- Detail Preservation: CLAHE tends to preserve more details in the image compared to Max Pooling. In some cases, Max Pooling can blur or eliminate fine details in the image.
